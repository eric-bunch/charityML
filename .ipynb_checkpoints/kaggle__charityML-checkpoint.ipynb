```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import itertools

# machine learning
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix

# jupyter sugar
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
pd.set_option('display.max_columns', None)
%matplotlib inline
```

# Load data

```python
census = pd.read_csv('census.csv')
```

```python
features = [c for c in census.columns if c != 'income']
target = 'income'
```

### Separate categorical and continuous features

```python
categorical_features = [
    'workclass',
    'education_level',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'native-country',
    'income'
]
```

```python
continuous_features = [
    'age',
    'education-num',
    'capital-gain',
    'capital-loss',
    'hours-per-week'
]
```

### Check for null values

```python
census.isnull().values.any()
```

### Pairplot of continuous features

```python
sns.pairplot(
    data=census,
    vars=continuous_features,
    size=3,
    plot_kws={'s': 7}
);
```

# Model

### Encoding dictionary

```python
cf_ledict = {
    feature: LabelEncoder().fit_transform(census[feature].values) for feature in categorical_features
}
```

```python
census_t = pd.merge(
    left=pd.DataFrame(cf_ledict).rename(columns={
        feature: feature + '_transformed' for feature in categorical_features
    }),
    right=census[continuous_features],
    right_index=True,
    left_index=True,
    copy=False
)
```

```python
features_t = [feature + '_transformed' for feature in categorical_features if feature != 'income'] + \
continuous_features
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(
    census_t[features_t].values,
    census_t[['income_transformed']],
    test_size=0.4
)
```

```python
tr = tree.DecisionTreeClassifier(
    criterion='gini',
    splitter='best',
    max_depth=10,
    max_leaf_nodes=None,
    class_weight=None
)
```

```python
tr.fit(
    X=X_train,
    y=y_train
)
```

## Model evaluation

#### Accuracy

```python
tr.score(X_test, y_test)
```

#### Precision and recall

```python
precision_recall_fscore_support(
    y_true=y_test,
    y_pred=tr.predict(X_test),
    average=None
)
```

#### Feature importances

```python
pd.DataFrame(
    data=tr.feature_importances_,
    index=features_t,
    columns=['feature_importances']
).sort_values('feature_importances', ascending=False)
```

#### Confusion matrix

```python
conf = pd.DataFrame(confusion_matrix(
    y_true=y_test,
    y_pred=tr.predict(X_test)
))
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

sns.heatmap(
    data=conf,
    ax=ax,
    annot=True,
    fmt='d'
);
```
